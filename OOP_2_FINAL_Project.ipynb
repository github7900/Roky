```python
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author


class Node:
    def __init__(self, book):
        self.book = book
        self.next = None


class Library:
    def __init__(self):
        self.head = None

    def create_node(self, book):
        return Node(book)

    def add_book(self, book):
        new_node = self.create_node(book)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node
        print("Book added successfully!")

    def remove_book(self, book_id):
        if self.head is None:
            print("No books found!")
            return
        if self.head.book.book_id == book_id:
            self.head = self.head.next
            print("Book removed successfully!")
            return
        temp = self.head
        while temp.next is not None and temp.next.book.book_id != book_id:
            temp = temp.next
        if temp.next is None:
            print(f"Book ID {book_id} not found!")
        else:
            temp.next = temp.next.next
            print("Book removed successfully!")

    def search_book(self, keyword):
        temp = self.head
        found = False
        while temp is not None:
            if temp.book.title == keyword or temp.book.author == keyword:
                print(f"Book ID: {temp.book.book_id}\nTitle: {temp.book.title}\nAuthor: {temp.book.author}\n")
                found = True
            temp = temp.next
        if not found:
            print("No books found with the given keyword!")

    def display_books(self):
        temp = self.head
        if temp is None:
            print("No books found!")
            return
        print("Book ID\tTitle\tAuthor")
        while temp is not None:
            print(f"{temp.book.book_id}\t{temp.book.title}\t{temp.book.author}")
            temp = temp.next

    def free_memory(self):
        self.head = None  # In Python, garbage collection handles memory deallocation.


def main():
    library = Library()
    while True:
        print("\n------ Library Management System ------")
        print("1. Add Book")
        print("2. Remove Book")
        print("3. Search Book")
        print("4. Display Books")
        print("5. Exit")
        choice = int(input("Enter your choice: "))

        if choice == 1:
            book_id = int(input("Enter book ID: "))
            title = input("Enter book title: ")
            author = input("Enter book author: ")
            book = Book(book_id, title, author)
            library.add_book(book)
        elif choice == 2:
            book_id = int(input("Enter book ID to remove: "))
            library.remove_book(book_id)
        elif choice == 3:
            keyword = input("Enter a keyword to search: ")
            library.search_book(keyword)
        elif choice == 4:
            library.display_books()
        elif choice == 5:
            print("Exiting program...")
            library.free_memory()
            break
        else:
            print("Invalid choice! Please try again.")


if __name__ == "__main__":
    main()
```
